### 比較演算子1

Pythonには2つの値を比較するときに使用する比較演算子が用意されています。

例えば、2つの値や変数が等しいかどうかを判定するには==を使います。

代入を表す=と区別するためにこのように表されます。

等しくないことを判断するには!=を用います。

比較した結果、正しければTrue、間違っていればFalseとなります。例えば、以下の例の結果は全てTrueになります。

    (1 + 1) == 2
    'hello' != 'python' 
    True != False
    'Python' != 'python'


### 問題1

    (1)サンプルを参考に、bool型のTrueとFalseが等しいか調べ、出力してみましょう。
    (2)サンプルを参考に、int型の-3 * 5と15は等しくないことを調べ、出力してみましょう。
    (3)サンプルを参考に、str型の'hello'と'HELLO'は等しいか調べ、出力してみましょう。
    (4)bool型とint型を比較するとどうなるでしょうか？サンプルを参考に、Trueと1は等しいか調べ、出力してみましょう。


```python
#問題1
#(サンプル) 2と3が等しいかを調べ、出力します。
print(2==3)
# TrueとFalseが等しいか調べ出力します。
print(True==False)
# -3 * 5と15は等しくないか調べ出力します。
print(-3*5==15)
# 'hello'と'HELLO'は等しいか調べ出力します。
print('hello'=='HELLO')
# Trueと1は等しいか調べ出力します。
print(True==1)
```


### ヒント

### 比較演算子2

Pythonにおいて、値の大小は>（大なり）と<（小なり）を使って表すことができます。

また、これらは等号を付け加えて>=（以上）、<=（以下）と表すこともできます。

<=は正しい文法ですが、=<は間違っていることに注意してください。

例えば、以下の例は全てTrueを返します。

    1 < 2
    1 <= 1
    'alpha' <= 'beta'





```python
print(1<2)
print(1<=1)
print('alpha'<='beta')
```

文字列の比較では、アルファベット順に基づいて大小関係が決定されます。

### 論理演算子(1)

bool型は1または0、TrueまたはFalseのいずれかです。

このbool型の演算を扱うために論理演算子が用意されています。

論理演算子and（かつ）、or（または）、not（否定）を使って論理演算を行うことができます。


image.png


例えば、以下の例は全てTrueを返します。

    True and True
    True or False
    not False


```python
print(True and True)
print(True or False)
print(not False)
```

もう少し具体的な例を見てみましょう。以下のような論理演算を考えます。

    a > 32 and b <= 50

この時a>32とb<=50の真偽がそれぞれ評価されて、どちらもTrueならばTrue、どちらか片方でもFalseならばFalseになります。

    a > 32 or b <= 50

この場合は、どちらか片方でもTrueならTrueとなり、どちらもFalseならFalseになります。

論理演算子を組み合わせることで、より高度な論理式を書くことができます。


### 問題3
    ソースコード欄には点数を表す2つの変数、my_scoreとyour_scoreが定義されています。
    以下を調べてprint関数で結果を出力してみましょう。

    (1)my_scoreは40より大きく、かつ44より小さい。
    (2)my_scoreは31より小さい、または41より大きい。
    (3)my_scoreの2倍はyour_scoreの3倍より小さい。

```python
#問題3
# 変数を定義しましょう
my_score = 44.0
your_score = 31.0
# my_scoreは40より大きく、かつ44より小さい
print(my_score > 40 and my_score < 44)
# my_scoreは31より小さい、または41より大きい
print(my_score < 31 or my_score > 41)
# my_scoreの2倍はyour_scoreの3倍より小さい
print(my_score * 2 < your_score * 3)
```


##### ヒント

- my_score > 40 and .... 空欄を埋めてください。
- my_score < 31 or .... 空欄を埋めてください。
- my_score * 2 < .... 空欄を埋めてください。

## if文

if文はプログラミングの基本的な考え方の一つです。if文によって、ある条件に当てはまる時だけ処理を実行することができます。

if文の基本的な書き方は、

    if a == b:
      print(a)

というように、if 条件:(コロン)の後改行して' 　'(タブスペース)の後に条件に合致した時の処理を書きます。

image.png

このように空けたタブスペースをインデントと呼び、Pythonでは下のコードのようにif文の中身などの処理のまとまりごとにインデントを分けなければいけません。

    if a == b:
      print(a)

    if c == d:
      print(c)

### 問題4

    あなたは数学のテストの採点官です。
    ソースコード欄にはテストの点数として変数pointsが書かれています。
    この点数に応じて何らかのコメントを出すため、
    if文を使ってプログラミングするとします。
    点数が60点以上の場合、"よくできました"と表示されるようにif文を書いてみましょう。


```python
#問題4
# 数学の点数
points = 70.0
# 点数に応じてコメントを出すif文を記述しましょう
if points >= 60:
    print('よくできました')
```


#### ヒント

- pointsが60以上である、というのはpoints >= 60と表せます。

- 条件を書いたあとに：（コロン）を忘れないようにしましょう。

- if文より下の文(if文の条件下で実行される文)には最初に' 　'(インデント)があります。

### elseの追加

else文を使うと、if文の条件がFalseだった場合の処理を記述することができます。

image.png

例えば、else文は下のように記述することができます。

    if a == b:
      print(a)
    else:
      print(b)

if文に対応するようにelse:を記述することで、それ以外のときの処理を記述することができます。

上の例では、aとbの値が一致しているときはaの値が表示され、それ以外のときはbの値が表示されます。

### 問題5

    点数pointsに50が代入されています。
    点数が60点以上の場合に"よくできました"、それ以外の場合に"頑張りましょう"と
    表示されるようにプログラムを作りましょう。


```python
#問題5
# 数学の点数
points = 50.0
# 点数に応じてコメントを出すif/else文を記述しましょう
if points >= 60:
    print('よくできました')
else:
    print('頑張りましょう')
```

### ヒント

- elseのあとは：（コロン）のみです。

- else文より下の文(else文の条件下で実行される文)には最初に' 　'(インデント)があります。

### さらにelifの追加

if/else文の組み合わせでは、ある条件か否かで処理を分けることができました。さらにelif文を使うことで、より複雑な条件分けをして処理を分けることができます。


image.png

例えば、elif文は下のように記述することができます。

    if a == b:
        print(a)
    elif a > b:
        print(b)
    else:
        print(“other”)

if文に対応するようにelif文を記述することでさらに条件分岐をすることができます。

上の例では、aとbの値が一致しているときはaの値が表示され、aの方がbよりも大きい値のときはbの値が表示されます。またどちらの条件にもあてはまらない場合、otherという文字列が表示されます。


### 問題6

    前回の課題では、点数が60点以上かどうかで、"よくできました"
    か"頑張りましょう"のどちらかのコメントを表示させました。
    今回はelif文を使って、さらに細かい条件設定をして、表示でき
    るコメントを増やしてみましょう。

    ・点数が50点以上の場合、"まあまあです"と表示されるようにelif文を書いてみましょう。


```python
#問題6
# 数学の点数
points = 50.0
# 点数に応じてコメントを出すif/elif/else文を記述しましょう
if points >= 60:
    print("よくできました")
elif points >= 50:
    print("まあまあです")
else:
    print("頑張りましょう")
```

### ヒント

- elifの後に条件を書きましょう。：（コロン）を忘れずに。

- elif文より下の文(elif文の条件下で実行される文)には最初に' 　'(インデント)があります。
